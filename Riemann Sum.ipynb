## Riemann Sum Funtion

#### A Riemann sum is an estimation of the area under a curve using rectangles. The right Riemann sum Rn of a function f on the interval [a, b] with n rectangles is calculated by Xn                                                                                                                                                  Rn = f(a + Δx)Δx + f(a + 2Δx)Δx + · · · + f(a + nΔx)Δx = f(a + kΔx)Δx, k=1                                                                                                             where Δx = b−a , is the width of the rectangles. Write a Python function to take in a mathematical function n f, two interval endpoints a and b, and the number of rectangles n, and then outputs the right Riemann sum Rn

```python
import numpy as np
import matplotlib.pyplot as plt

def my_f(x):
    return eval(input()) # evaluates the input function

def my_func(f,a,b,n):
    x = np.linspace(a,b,n+1)
    dx = (b-a)/n
    x_right = np.linspace(dx,b,n) # Right Riemann Sum formula
    rn = np.sum(f(x_right)*dx)
    return rn

print ("The right Riemann sum of the function is :" ,(my_func((my_f),0,5,10)))
```

### Using the Right Riemann Sum to Find the Area Under a Curve
#### Let f(x) = 4 − x2

##### (a) Sketch a graph of f and the rectangles that make up the right Riemann sum R4 on the interval [0, 2].This can be done by hand

```python
import numpy as np
import matplotlib.pyplot as plt
f = lambda x: (4-(x**2))**1/2 # lamda argument: expression
x = np.linspace(0,2,4+1)
y = f(x)


plt.plot( x,y)
x_right = x[1:] # right end points
y_right = y[1:]
plt.plot(x_right,y_right,'b.',markersize=10)
plt.bar(x_right, y_right, width=-(2-0)/4, alpha=0.2, align='edge', edgecolor='b') # plot bar with align right-edge
plt.title('Right Riemann Sum, n = 4')

plt.show()


```
